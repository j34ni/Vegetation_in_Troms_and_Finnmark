# Merge yearly .hdf files into input & output .csv files
## Do not use 2019 (since the output is all null)

```python
print('Imports')
import glob
import os
import pandas as pd
```

```python
# Input to .csv
print('Input files')
path = '/home/jovyan/Arctic/Vegetation_in_Troms_and_Finnmark/data/'
x_filename = 'x_mean_tp_2*.hdf'
file_list = glob.glob(os.path.join(path, x_filename))
file_list.sort()
file_list

input_filename = os.path.join(path, 'input_mean_tp.csv')

for file_name in file_list[0:-1]:
    print(file_name)
    df = pd.read_hdf(file_name)
    print(df)
    df = df.drop(columns=['lon_lat'])
    if (file_name == file_list[0]):
         df.to_csv(input_filename, header=True, index=None, sep=',')
    else:
        # Append to file
         df.to_csv(input_filename, mode='a', header=None, index=None, sep=',')
```

```python
# Output to .csv
print('Output files')
y_filename = 'y_mean_tp_2*.hdf'
file_list = glob.glob(os.path.join(path, y_filename))
file_list.sort()
file_list

output_filename = os.path.join(path, 'output_mean_tp.csv')

for file_name in file_list[0:-1]:
    print(file_name)
    dg = pd.read_hdf(file_name)
    print(dg)
    dg = dg.reset_index()
    dg = dg.drop(columns=['lon_lat'])
    if (file_name == file_list[0]):
         dg.to_csv(output_filename, header=True, index=None, sep=',')
    else:
        # Append to file
         dg.to_csv(output_filename, mode='a',  header=None, index=None, sep=',')
```
